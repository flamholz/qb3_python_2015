```python
import csv

def get_gene_map(genome_info_fname):
    """Reads in the genome annotation data.
    
    Args:
        genome_info_fname: filename for the tab-delimited genome
            info file.
    
    Returns:
        A dictionary mapping the locusId to a dictionary of 
        gene annotations.
    """
    genes_data = {}
    with open(genome_info_fname) as inf:
        # Dict reader reads a tabular file and returns
        # each row as a dictionary mapping rows headers to values
        # for that row.
        reader = csv.DictReader(inf, dialect='excel-tab')
        for row in reader:
            # Store the gene info by locus ID genes_data.
            gene_name = row['locusId']
            genes_data[gene_name] = row
    # genes_data is now initialized.
    return genes_data


def get_operon_list(operon_fname):
    """Reads in the operon annotation data.
    
    Args:
        operon_pred_fname: filename for operon predictions.
    
    Returns:
        A list of sets of locus IDs of all operons containing 
        more than 1 gene.
    """
    # General strategy: Keep a dictionary mapping locus IDs to sets of genes
    # in that operon. If two genes are in the same operon, they will have the 
    # same set in the dictionary. This allows us to find the operon from
    # the locus ID of any gene that is already in it.
    operon_map = {}  # maps gene_names to indexes in operon_list
    operon_list = [] # keep a list of the unique operons for returning
    with open(operon_fname) as inf:
        # Another DictReader, like above.
        reader = csv.DictReader(inf, dialect='excel-tab')
        
        operon_genes = set()
        for row in reader:
            gene_a = row['Gene1'].strip()
            gene_b = row['Gene2'].strip()
            # Names are valid if both evaluate to True, i.e. neither is empty.
            valid_names = gene_a and gene_b
            # bOp is a prediction of whether both are in an operon
            in_operon = (row['bOp'] == 'TRUE')  
            
            # Consider something in an operon only if both 
            # genes have valid (non-empty) names. 
            # Since we use the locus ID for the name, I believe all of them 
            # are valid. But this is called "defensive programming."
            in_operon = in_operon and valid_names
                        
            # Current pair not in an operon
            if in_operon:
                if gene_a in operon_map:
                    operon_map[gene_a].add(gene_b)
                    operon_map[gene_b] = operon_map[gene_a]
                else:
                    s = set([gene_a, gene_b])
                    operon_list.append(s)
                    operon_map[gene_a] = s
                    operon_map[gene_b] = s

    return operon_list 


# Template for an operon row.
OPERON_ROW_KEYS = ['seqname', 'source', 'feature', 'start',
                   'end', 'score', 'strand', 'frame', 'attributes']
OPERON_ROW_TEMPLATE = {'seqname': 'Chromosome',
                       'source': 'MicrobesOnline',
                       'feature': 'operon',
                       'score': '.',
                       'frame': '.'}
def make_operon_row(strand, start, stop, attributes):
    """Returns a dictionary representing the operon."""
    # Copy the default operon row and modify.
    op_data = OPERON_ROW_TEMPLATE.copy()    
    op_data['strand'] = strand
    op_data['start'] = start
    op_data['end'] = stop
    op_data['attributes'] = attributes
    return op_data
```

```python
gene_map = get_gene_map('genomeInfo.tsv')
operon_list = get_operon_list('gnc511145.named')
all_genes_in_operons = set()
for op in operon_list:
    all_genes_in_operons.update(op)

# Make this GFF file now. 
# This first loop only covers genes in multi-gene operons.
operon_rows = []
for op in operon_list:
    names = [gene_map[n]['name'] for n in op]
    op_name = '-'.join(sorted(names))
    strands = set([gene_map[n]['strand'] for n in op])
    if len(strands) != 1:
        print 'Operon genes not on same strand, skipping'
        continue
        
    strand = strands.pop()
    
    starts = [int(gene_map[n]['start']) for n in op]
    stops = [int(gene_map[n]['stop']) for n in op]
    if strand == '+':
        # If we are on the + strand, the start of the operon
        # is the first start position of any gene and the end
        # is the last stop of any gene.
        start = min(starts)
        stop = max(stops)
    else:
        # If we are on the - strand, numbers are reverse.
        start = max(starts)
        stop = min(stops)
    
    attr = r'operon "%s"' % op_name 
    op_data = make_operon_row(strand, start, stop, attr)
    operon_rows.append(op_data)

    
# Notice that we omitted genes that are not in operons.
# Let's make operons for all the genes we missed.
all_genes = set(gene_map.keys())
genes_not_in_operons = all_genes.difference(all_genes_in_operons)
for gene in genes_not_in_operons:
    d = gene_map[gene]
    start = int(d['start'])
    stop = int(d['stop'])
    strand = d['strand']
    attributes = r'operon "%s"' % d['name']
    op_data = make_operon_row(strand, start, stop, attr)
    operon_rows.append(op_data)

# Order in which to write the fields to the file. 
with open('GFFoutput.gff', 'w') as outf:
    # OPERON_ROW_KEYS is defined above and determines the order
    # in which the DictWriter writes the fields in the dictionary.
    writer = csv.DictWriter(outf, OPERON_ROW_KEYS, dialect='excel-tab',
                            quoting=csv.QUOTE_NONE,
                            escapechar='\\')
    writer.writerows(operon_rows)  # Writes all the rows.
```
